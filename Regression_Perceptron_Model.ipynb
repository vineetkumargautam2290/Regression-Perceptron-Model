# Regression Problem

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv("/content/drive/MyDrive/Udemy data science/Allfiles/boston.csv",index_col=0) # importing data
data.head()
```

```python
data.isnull().sum()
```

No NUll values

```python
# defining independent and dependent variables
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x.head()
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
pd.DataFrame(x).head()
```

```python
y.head()
```

```python
# dividing dataset into traning and testing datasets
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape )
```

# MultiLayer Perceptron Model - DNN

```python
# making a deep learning model
dnn = Sequential()
# add hidden layer 1
dnn.add(Dense(16, activation='relu', input_dim = 13))
# batch normalization
#dnn.add(BatchNormalization())
# dropout
#dnn.add(Dropout(0.25))
# add one more hidden layer
#dnn.add(Dense(8, activation='relu'))
# batch normalization
#dnn.add(BatchNormalization())
# dropout
#dnn.add(Dropout(0.5))
# output
dnn.add(Dense(1, activation='linear'))
# summary
dnn.summary()
```

```python
# compile the model while using optimization and loss function
dnn.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model
history = dnn.fit(x_train, y_train, epochs=500)
```

```python
# Predict the test dataset
y_pred = dnn.predict(x_test)
y_pred
```

```python
print(y_pred[0:10])
print("************"*5)
print(y_test[0:10])
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test, y_pred))
```
